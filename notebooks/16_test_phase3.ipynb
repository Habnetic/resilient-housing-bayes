```python
import sys
from pathlib import Path

# add src/ to python path
sys.path.append(str(Path().resolve().parent / "src"))

import pandas as pd
from rhb.transfer_scaling import fit_reference_scaler, apply_reference_scaler, derive_e_hat_v0

df_rtm = pd.DataFrame({
    "dist_to_water_m": [10, 20, 30],
    "water_len_density_250m": [1, 2, 3],
    "water_len_density_500m": [2, 3, 4],
    "water_len_density_1000m": [5, 6, 7],
})

df_ham = pd.DataFrame({
    "dist_to_water_m": [15, 25],
    "water_len_density_250m": [1.5, 2.5],
    "water_len_density_500m": [2.5, 3.5],
    "water_len_density_1000m": [5.5, 6.5],
})

cols = [
    "dist_to_water_m",
    "water_len_density_250m",
    "water_len_density_500m",
    "water_len_density_1000m",
]

scaler = fit_reference_scaler(df_rtm, cols)
df_ham = apply_reference_scaler(df_ham, scaler)
df_ham = derive_e_hat_v0(df_ham)

print(df_ham)
```
